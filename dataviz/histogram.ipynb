```python
# loading the package
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\t480\Desktop\S2\technohack\Iris.csv") # loading the dataset
```

```python
df_columns = df.columns.tolist() # extracting the columns of the datase as list
df_columns 
```

# DATA VIZ USING MATPLOTLIB 
1- HISTOGRAM PLOT WITHOUT NORMALISATION

```python
for column in df_columns: # looping througth each columns in the dataframe
    if column != 'Species' and column != 'Id': # making sure to plot just the target columns
        sns.set_style('white') # make the plot backgraod as white 
        plt.figure(figsize=(8,6)) # set the size of the plot figure
        plt.hist(df[column], bins = 7, color= '#108A99') # set the plot type as histogram white 7 bars and colore
        if column != 'SepalWidthCm' or column != 'PetalWidthCm':
            plt.title(f'Distribution of {column[0:5]} {column[5:10]} in Cm', fontsize= 14, weight = 'bold') # here we set the titel of the plot with its size and weight
            plt.ylabel(f'The count of {column[0:5]} {column[5:10]}') # set the name of the y labe
            plt.xlabel(f'{column[0:5]} {column[5:10]} in Cm')
        else:
            plt.ylabel(f'The count of {column[0:5]} {column[5:10]}') # set the name of the y labe
            plt.xlabel(f'{column[0:5]} {column[5:11]} in Cm') # the x label 
            plt.xlabel(f'{column[0:5]} {column[5:11]} in Cm') # set the name of xlabel
        sns.despine() # try yourself to figure out the role of this parametre
        plt.show()
```

2- HISTOGRAM PLOT WITH NORMALISATION OF VALUES

```python
df_new = pd.DataFrame(columns=df.columns) # create new dataset 
for column in df.columns:# looping througth each column of dataste
    if column != 'Species' and column != 'Id': # do the operation to all columns except column Id and Species
        sigma = df[column].std() # calculat the standard diviation for each column
        mu = df[column].mean() # calculat the mean for each column
        df_new[column] = (df[column] - mu) / sigma   # modifie each columns
df_new.head()
```

```python
for column in df_new.columns:
    if column != 'Species' and column != 'Id':
        sns.set_style('white') # make the plot bachgraod as white 
        plt.figure(figsize=(8,6)) # set the size of the plot figure
        plt.hist(df[column], bins = 7, color= '#108A99', density=True) # set the plot type as histogram white 7 bars and colore
        if column != 'SepalWidthCm' or column != 'PetalWidthCm':
            plt.title(f'Distribution of {column[0:5]} {column[5:10]} in Cm', fontsize= 14, weight = 'bold') # here we set the titel of the plot with its size and weight
        else:
             plt.title(f'Distribution of {column[0:5]} {column[5:11]} in Cm', fontsize= 14, weight = 'bold') # here we set the titel of the plot with its size and weight
        plt.ylabel('Probability density')
        plt.xlabel('The values')# set the name of xlabel
        sns.despine() # try yourself to figure out the role of this parametre
        plt.show()
```

2- HISPLOT

```python
# plot  all the columns
for column in df.columns: # looping througth each column of dataste
    if column!='Id' and column!='Species' : # macking sure that all the columns will be ploted except  Species and Id
        sns.set_style('white') # macking the background white
        sns.histplot(df, x= column, kde=True) # usinf histplot of seaborn 
        sns.despine() 
        plt.show()
```

```python
# for all the columns in concidiration of Species
for column in df.columns: # looping througth each column of dataste
    if column!='Id' and column!='Species' :  # macking sure that all the columns will be ploted except  Species and Id
        sns.set_style('white') # macking the background white
        sns.histplot(df, x= column, hue='Species', kde=True)# using histplot of seaborn to plot all Species 
        sns.despine()
        plt.show()
```
